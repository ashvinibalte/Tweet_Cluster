```python
import snscrape.modules.twitter as sntwitter
import pandas as pd

query = "HPV vaccine"
tweets = []
limit = 100000


for tweet in sntwitter.TwitterSearchScraper(query).get_items():

    # print(vars(tweet))
    # break
    if len(tweets) == limit:
        break
    else:
        tweets.append([tweet.date, tweet.username, tweet.content])

df = pd.DataFrame(tweets, columns=['Date', 'User', 'Tweet'])
print(df)

# to save to csv
df.to_csv('HPV Vaccine.csv')
```

```python
df = df.drop_duplicates('Tweet')
```

```python
import pandas as pd
import re
from bs4 import BeautifulSoup
import string

# Remove URLs from a specific column
df['Tweet'] = df['Tweet'].apply(lambda x: re.sub(r'http\S+', '', x))

# Remove HTML tags from a specific column
df['Tweet'] = df['Tweet'].apply(lambda x: BeautifulSoup(x, 'lxml').get_text())

# Remove punctuation from a specific column
df['Tweet'] = df['Tweet'].apply(lambda x: x.translate(str.maketrans('', '', string.punctuation)))

# Remove smileys from a specific column
df['Tweet'] = df['Tweet'].apply(lambda x: re.sub('[^a-zA-Z0-9\s]+', '', x))

# Convert Text in a specific column to lowercase
df['Tweet'] = df['Tweet'].str.lower()

# Remove newline characters from a specific column
df['Tweet'] = df['Tweet'].str.replace('\n', '')

# Remove "@" symbol from a specific column
df['Tweet'] = df['Tweet'].str.replace('@', '')
```

```python
import langid

# Identify the language of each row in a specific column
df['language'] = df['Tweet'].apply(lambda x: langid.classify(x)[0])

# Keep only the rows where the language is English
df = df[df['language'] == 'en']

# Remove the language column
df = df.drop(columns=['language'])
```
